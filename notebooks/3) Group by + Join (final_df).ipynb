```python
import pandas as pd

def left_merge (df1,df2,column):
    left_merge =df1.merge(df2, how='left', on = [column])
    return left_merge

def join_col_values (df, new_col, col1, col2, sep):
    df['new_col'] = df['col1'].str.cat(df['col2'],sep="sep")
    return df

def inner_merge (df1,df2,column):
    inner_merge = df1.merge(df2, how='inner', on=[column])
    return inner_merge
```

```python
student_info = pd.read_csv("/Users/stefaniabraca/Desktop/final_project/data/raw/studentInfo.csv")
student_asmt = pd.read_csv("/Users/stefaniabraca/Desktop/final_project/data/raw/studentAssessment.csv")
student_vle = pd.read_csv("/Users/stefaniabraca/Desktop/final_project/data/raw/studentVle.csv")
```

### student_asmt  groupby

```python
assignments_agg = student_asmt.groupby('id_student').agg(mean_score=('score', 'mean'), 
                                                        n_assignments=('score', 'count'))
assignments_agg
```

### student_vle groupby

```python
activity_agg = student_vle.groupby('id_student').agg(total_clicks=('sum_click', 'sum'))
activity_agg
```

### Join

```python
first_join = inner_merge(student_info, assignments_agg, "id_student")
```

```python
second_join = inner_merge(first_join, activity_agg, "id_student")
```

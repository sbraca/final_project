```python
import pandas as pd
import numpy as np
```

```python
six_clusters = pd.read_csv("../data/processed/5_clusters_index_df.csv")
```

```python
six_clusters.head()
```

```python
clusters_agg = six_clusters.groupby('cluster_result').agg(mean_score=('mean_score', 'mean'), 
                                                    mean_asmt=('n_assignments', 'mean'), 
                                                    mean_clicks=('total_clicks', 'mean'), 
                                                    mean_result=('final_result', 'mean'), 
                                                    id_count=('id_student', 'count'),
                                                    region_mode=('region', pd.Series.mode))
```

```python
import pandas as pd
import numpy as np
```

```python
clusters_df = pd.read_csv("../data/processed/clusters_df.csv")
```

```python
clustering= pd.read_csv("../data/processed/final_clustering.csv")
```

```python
clusters_agg = clusters_df.groupby('cluster_1').agg(mean_score=('mean_score', 'mean'), 
                                                    mean_asmt=('n_assignments', 'mean'), 
                                                    mean_clicks=('total_clicks', 'mean'), 
                                                    mean_result=('final_result', 'mean'), 
                                                    id_count=('final_result', 'count'),
                                                    region_mode=('region', pd.Series.mode))
```

```python
clusters_agg
```

```python
clustering
```

```python
clustering['final_result'] = clustering['final_result'].apply(lambda x: str(x).replace('Distinction','4'))
clustering['final_result'] = clustering['final_result'].apply(lambda x: str(x).replace('Pass','3'))
clustering['final_result'] = clustering['final_result'].apply(lambda x: str(x).replace('Fail','2'))
clustering['final_result'] = clustering['final_result'].apply(lambda x: str(x).replace('Withdrawn','1'))
```

```python
clustering
```

```python
clustering_agg = clustering.groupby('cluster').agg(mean_score=('mean_score', 'mean'), 
                                                    mean_asmt=('n_assignments', 'mean'), 
                                                    mean_clicks=('total_clicks', 'mean'), 
                                                    mean_result=('final_result', 'mean'), 
                                                    id_count=('final_result', 'count'))
```

```python
import pandas as pd
import numpy as np
```

```python
clusters_df = pd.read_csv("../data/processed/clusters_df.csv")
```

```python
clusters_df
```

```python
clusters_agg = clusters_df.groupby('cluster_1').agg(mean_score=('mean_score', 'mean'), 
                                                    mean_asmt=('n_assignments', 'mean'), 
                                                    mean_clicks=('total_clicks', 'mean'), 
                                                    mean_result=('final_result', 'mean'), 
                                                    id_count=('final_result', 'count'),
                                                    region_mode=('region', pd.Series.mode))
```

```python
import pandas as pd
import numpy as np
```

```python
HDBSCAN_7 = pd.read_csv("../data/processed/0.25_seven_clusters_df.csv")
```

```python
 DBSCAN_8 = pd.read_csv("../data/processed/DBSCANclusters_index_df.csv")
```

```python
HDBSCAN_7.head()
```

```python
HDBSCAN_7.cluster_result.unique()
```

```python
HDBSCAN_7['final_result'] = HDBSCAN_7['final_result'].apply(lambda x: str(x).replace('Distinction','4'))
HDBSCAN_7['final_result'] = HDBSCAN_7['final_result'].apply(lambda x: str(x).replace('Pass','3'))
HDBSCAN_7['final_result'] = HDBSCAN_7['final_result'].apply(lambda x: str(x).replace('Fail','2'))
HDBSCAN_7['final_result'] = HDBSCAN_7['final_result'].apply(lambda x: str(x).replace('Withdrawn','1'))
```

```python
HDBSCAN_7.dtypes
```

```python
HDBSCAN_7[["final_result"]]=HDBSCAN_7[["final_result"]].apply(pd.to_numeric)
```

```python
HDBSCAN_7 = HDBSCAN_7.groupby('cluster_result').agg(mean_score=('mean_score', 'mean'), 
                                                    mean_asmt=('n_assignments', 'mean'), 
                                                    mean_clicks=('total_clicks', 'mean'), 
                                                    mean_result=('final_result', 'mean'), 
                                                    id_count=('final_result', 'count'),
                                                    region_mode=('region', pd.Series.mode))
```
